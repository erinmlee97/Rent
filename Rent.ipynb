```python
# Dependencies
import pandas as pd
from opencage.geocoder import OpenCageGeocode
from config import api_key
```

```python
# Read csv file into pandasdf.drop(columns=['B', 'C'])
rent = pd.read_csv("Resources/rent.csv")
rent
```

```python
rent = rent.drop(columns=['SizeRank', 'RegionID'])
rent.drop(rent.index[0])
```

```python
rent[['City','State']] = rent['RegionName'].str.split(',',expand=True)
rent
```

```python
geocoder = OpenCageGeocode(api_key)
```

```python
# Create list to hold coordinates
list_lat = []

list_long = []

# Iterate over rows in dataframe
for index, row in rent.iterrows(): 
    City = row['City']
    State = row['State']       
    query = str(City)+','+str(State)

    results = geocoder.geocode(query)   
    lat = results[0]['geometry']['lat']
    long = results[0]['geometry']['lng']

    list_lat.append(lat)
    list_long.append(long)

# Create new columns from lists    
rent['lat'] = list_lat   
rent['lon'] = list_long

rent
```

```python
# Rearrange Columns
cols = rent.columns.tolist()
cols.insert(0, cols.pop(cols.index('RegionName')))
cols.insert(1, cols.pop(cols.index('City')))
cols.insert(2, cols.pop(cols.index('State')))
cols.insert(3, cols.pop(cols.index('lat')))
cols.insert(4, cols.pop(cols.index('lon')))

# Reindex DataFrame
rent = rent.reindex(columns= cols)

rent
```

```python
rent = rent.melt(id_vars=["RegionName", "City", "State", "lat", "lon"], 
        var_name="Date", 
        value_name="Rent")
```

```python
rent = rent.sort_values(["RegionName", "City", "State", "lat", "lon"])
rent
```

```python
rent.to_csv('Resources/rent_cleaned.csv')
```
